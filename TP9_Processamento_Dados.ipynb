## Criação de arquivos JSON

```python
import json
import csv

# Nome do arquivo CSV de entrada
csv_Chamados = 'arquivos_csv/chamados.csv'
csv_Usuarios = 'arquivos_csv/usuarios.csv'
csv_Departamentos = 'arquivos_csv/departamentos.csv'
csv_Inventario = 'arquivos_csv/inventario.csv'

# Nome do arquivo JSON de saída
json_Chamados = 'arquivos_json/chamados.json'
json_Usuarios = 'arquivos_json/usuarios.json'
json_Departamentos = 'arquivos_json/departamentos.json'
json_Inventario = 'arquivos_json/inventario.json'

# Lista para armazenar os dados do CSV
dados_csv_Chamados = []
dados_csv_Usuarios = []
dados_csv_Departamentos = []
dados_csv_Inventario = []

# Abrir arquivo no modo leitura e incluir dados nas listas acima
with open(csv_Chamados, mode='r', newline='', encoding='utf-8') as arquivo:
    leitor_csv = csv.DictReader(arquivo)
    for linha in leitor_csv:
        dados_csv_Chamados.append(linha) 

with open(csv_Usuarios, mode='r', newline='', encoding='utf-8') as arquivo:
    leitor_csv = csv.DictReader(arquivo)
    for linha in leitor_csv:
        dados_csv_Usuarios.append(linha)
        
with open(csv_Departamentos, mode='r', newline='', encoding='utf-8') as arquivo:
    leitor_csv = csv.DictReader(arquivo)
    for linha in leitor_csv:
        dados_csv_Departamentos.append(linha)

with open(csv_Inventario, mode='r', newline='', encoding='utf-8') as arquivo:
    leitor_csv = csv.DictReader(arquivo)
    for linha in leitor_csv:
        dados_csv_Inventario.append(linha)
        

# Escrever os dados em um arquivo JSON (Apenas para salvar)
with open(json_Chamados, 'w', encoding='utf-8') as arquivo:
    json.dump(dados_csv_Chamados, arquivo, indent=4, ensure_ascii=False)
    
with open(json_Usuarios, 'w', encoding='utf-8') as arquivo:
    json.dump(dados_csv_Usuarios, arquivo, indent=4, ensure_ascii=False)
    
with open(json_Departamentos, 'w', encoding='utf-8') as arquivo:
    json.dump(dados_csv_Departamentos, arquivo, indent=4, ensure_ascii=False)
    
with open(json_Inventario, 'w', encoding='utf-8') as arquivo:
    json.dump(dados_csv_Inventario, arquivo, indent=4, ensure_ascii=False)
```

```python
# Conexão com banco de dados

conn = sqlite3.connect('chamados.db')
```

```python
# Criação da tabela chamados

create_chamados_table_sql = '''
create table if not exists chamados_TI (
    ID int primary key,
    titulo text,
    Descricao text,
    Data_abertura text,
    Status text,
    Prioridade int,
    ID_Departamento int,
    ID_Usuario int
)
'''
conn.execute(create_chamados_table_sql)
```

```python
# Utilizando pandas para inserir os registros na tabela
dfChamados.to_sql('chamados_TI', conn, if_exists = 'replace', index = False)
```

```python
# Criando tabelas para Usuarios e Departamentos

create_usuarios_table_sql = '''
create table if not exists Usuarios (
    ID int primary key,
    Nome text,
    Email text,
    Id_Departamento int
)
'''

create_departamentos_table_sql = '''
create table if not exists Departamentos (
    ID int primary key,
    Departamentos text
)
'''

conn.execute(create_usuarios_table_sql)
conn.execute(create_departamentos_table_sql)
```

```python
dfUsuarios.to_sql('Usuarios', conn, if_exists = 'replace', index = False)
dfDepartamentos.to_sql('Departamentos', conn, if_exists = 'replace', index = False)
```

### Relação de todos os usuários por departamento em ordem alfabética.

```python
query1 = '''
select 
    Usuarios.Nome, 
    Departamentos.Departamentos
from Usuarios
join Departamentos on Usuarios.ID_Departamento = Departamentos.id
order by Departamentos.Departamentos desc, Usuarios.Nome asc
'''

resultado_query1 = pd.read_sql_query(query1, conn)
resultado_query1
```

### Relação de todos os chamados por grau de prioridade de atendimento (baixa, normal ou alta).

```python
query2 = '''
select
    Chamados_ti.id,
    Chamados_ti.titulo,
    Chamados_ti.descricao,
    Chamados_ti.data_abertura,
    Chamados_ti.status,
    Chamados_ti.prioridade,
    Departamentos.Departamentos as Departamento,
    Usuarios.Nome as Usuario
from Chamados_ti
join Departamentos on Chamados_ti.ID_Departamento = Departamentos.ID
join Usuarios on Chamados_ti.ID_Usuario = Usuarios.ID
order by Prioridade asc
'''
resultado_query2 = pd.read_sql_query(query2, conn)

# Função para mudar os valores de "Prioridade"
def muda_valor(x):
    if x == 1:
        return "Baixa"
    elif x == 2:
        return "Média"
    elif x == 3:
        return "Alta"

resultado_query2['Prioridade'] = resultado_query2['Prioridade'].map(muda_valor)
```

### Quem foi o usuário que abriu mais chamados e em que quantidade?

```python
query3 = '''
select
    Usuarios.nome as Usuario,
    Departamentos.Departamentos as Departamento,
    count(Chamados_ti.ID) as Total_Chamados
from Usuarios
join Chamados_ti on Usuarios.id = Chamados_ti.ID_Usuario
join Departamentos on Usuarios.ID_Departamento = Departamentos.ID
group by Usuarios.ID, Usuarios.Nome, Departamentos.Departamentos
order by Total_Chamados desc
limit 1
'''

resultado_query3 = pd.read_sql_query(query3, conn)
resultado_query3
```

### Quantidade de chamados por departamento.

```python
query4 = '''
select
    Departamentos.Departamentos as DepartamentoS,
    count(chamados_ti.ID) as Quantidade_Chamados
from chamados_ti
join Usuarios on chamados_ti.ID_Usuario = Usuarios.ID
join Departamentos on Usuarios.ID_Departamento = Departamentos.ID
group by Departamentos.ID, Departamentos.Departamentos
order by Quantidade_Chamados asc
'''
resultado_query4 = pd.read_sql_query(query4, conn)
resultado_query4
```
